```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = None

while True:
    print("\n========== GLOBAL HAPPINESS ANALYSIS ==========")
    print("1. Load Dataset")
    print("2. Show First 5 Rows")
    print("3. Top 10 Happiest Countries")
    print("4. GDP vs Happiness Graph")
    print("5. Correlation Heatmap")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        try:
            df = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\My Python Project\Global Happiness Report Analysis\happiness.csv")
            print("\n✅ Dataset Loaded Successfully!\n")
        except:
            print("\n❌ Error: File not found!")

    elif choice == '2':
        if df is not None:
            print("\n📊 First 5 Rows:\n")
            print(df.head())
        else:
            print("⚠️ Load dataset first!")

    elif choice == '3':
        if df is not None:
            top = df.sort_values(by="Happiness Score", ascending=False).head(10)
            print("\n🏆 Top 10 Happiest Countries:\n")
            print(top[['Country','Happiness Score']])

            plt.figure()
            sns.barplot(x="Happiness Score", y="Country", data=top)
            plt.title("Top 10 Happiest Countries")
            plt.show()
        else:
            print("⚠️ Load dataset first!")

    elif choice == '4':
        if df is not None:
            plt.figure()
            sns.scatterplot(x="GDP per capita", y="Happiness Score", data=df)
            plt.title("GDP vs Happiness")
            plt.show()
        else:
            print("⚠️ Load dataset first!")

    elif choice == '5':
        if df is not None:
            plt.figure()
            numeric_df = df.select_dtypes(include=['number'])
            sns.heatmap(numeric_df.corr(), annot=True)
            plt.title("Correlation Heatmap")
            plt.show()
        else:
            print("⚠️ Load dataset first!")

    elif choice == '6':   # ✅ Correct alignment
        print("👋 Exit Program")
        break

    else:   # ✅ Correct alignment
        print("❌ Invalid choice")
```
